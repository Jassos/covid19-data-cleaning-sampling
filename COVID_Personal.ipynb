## Reestructuración de DataSet 2024

Notando que el conjunto de datos de registros del 2024 contiene variables y datos que no se contemplan en los años anteriores, se planteó reestructurar los mismos datos y las columnas con el fin de:

1. No perder una gran cantidad de datos importantes.
2. Asegurar que el análisis estadístico posterior sea más congruente.


```python
df_2024 = df_2024[df_2024['RESULTADO_PCR'].isin([34, 5, 997, 999, 17])]
```

Se relacionan estos números preferentes **34, 5, 997, 999, 17** como el equivalente a los presentes en los demás DataSets.

Se renombran las columnas para poder interactuar con el restro de DataSet sin problemas, ya que se han editado las columnas para que los datos coincidan numerica y estadisticamente.

```python
df_2024 = df_2024.rename(columns={'RESULTADO_PCR': 'RESULTADO_LAB', 'CLASIFICACION_FINAL_COVID': 'CLASIFICACION_FINAL'})

df_2024 = df_2024.drop(columns=['RESULTADO_PCR_COINFECCION', 'CLASIFICACION_FINAL_FLU'])
```

```python
mapeo_resultado_lab_CONVERSION = {
    34: 1,
    5: 2,
    999: 3,
    17: 4,
    997: 97
}

df_2024['RESULTADO_LAB'] = df_2024['RESULTADO_LAB'].map(mapeo_resultado_lab_CONVERSION)

df_2024['RESULTADO_LAB'].unique()
```
